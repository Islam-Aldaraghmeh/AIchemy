```python
import pandas as pd

# Open the file and process it line by line
with open('L2.csv', 'r') as f:
    # 1. Read all lines
    lines = f.readlines()

    # 2. Get the header (first line)
    header = lines[0].strip().split(';')

    # 3. Process the data (remaining lines)
    # The '3' in split tells Python to stop splitting after the 3rd semicolon,
    # keeping the complex SMILES string intact.
    data = [line.strip().split(';', 3) for line in lines[1:] if line.strip()]

# 4. Create the DataFrame directly from our list (Do NOT use pd.read_csv here)
df = pd.DataFrame(data, columns=header)

# Optional: Set the 'index' column as the actual dataframe index
if 'index' in df.columns:
    df = df.set_index('index')

# Check the output
print(df.head())
print("\nExample of complex SMILES data:")
print(df['SMILES'].iloc[0])
```
